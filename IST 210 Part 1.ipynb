Idea 1
Car dealerships have many moving parts to their business that should be analyzed to increase efficiency and therefore increase profits. On one hand you have the auto manufacturer that requires data from the dealerships and dealerships who they themselves collect data from the customer. This is unique because it doesn't just involve a direct collection of data between the seller and customer. It's a collection of data between multiple parties, all in which collects different information that is beneficial to each. In this database, we will be analyzing five seperate groups: The amount of cars designated to each dealership, types of models designated to the dealership, the model of car the dealership sells to the consumer, the price of the car sold, and how many cars are left for sale on the dealership lot. In the first table, "The amount of cars designated to each dealership", we will track how many cars the manufactures send to each dealer. In our next table, "Types of models designated to the dealership", we will analyze the differnt models that each dealership will recieve from the manufacturers. In addition our next table, "The model of car the dealership sells to the consumer" will track what model is actually sold to the consumer to help track which models are selling well and which ones aren't. For "The price of the vehicle sold", will track how much each car was sold for so it can be compared to the MSRP price set to the manufacturer. Lastly, "The amount of cars remaining at the dealership" will track how many vehicles are still for sale at the dealership so the manufacturer knows whether or not to supply more cars to the dealership.

Idea 2
Farmers are known to collect massive amounts of data in order to maximize crop production. This is imperative for any farmer that wants to yield the maximum number of crops per plot of land, and is a necessity for tracking customer purchases. This is unique because unlike other databases, farmers are able to input certain values for ceratin tables that will have direct effects on other data tables in the same database. In this database, we will be analyzing five separate groups: number of plots each with a specific plot number; what is grown or bred in each plot; number of pestacides and what pestacides are being used in each plot; output of crops per plot; and for an end-user data table, customer purchases. In the first table, "Number of plots each with a specific plot number," we will track how many plots of land there are and what their unique plot ID is. Next, "what is grown or bred in each plot" will include data about what type of animal or plant is being grown in each plot. "Number of pestacides and what pestacides are being used in each plot" will consist of the type of pestacide used in each plot and a measurement of the amount used. "Output of crops per plot" will keep track of the crop yield for each plot. Lastly, "customer purchases" will keep track of the type and amount of items a consumer purchases, and tracks the total amount paid. This information will greatly enhance and maximize the producctivity of the farm.
