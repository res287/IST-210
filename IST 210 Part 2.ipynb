Finalized Idea
Car dealerships have many moving parts to their business that should be analyzed to increase efficiency and therefore increase profits. On one hand you have the auto manufacturer that requires data from the dealerships and dealerships who they themselves collect data from the customer. This is unique because it doesn't just involve a direct collection of data between the seller and customer. It's a collection of data between multiple parties, all in which collects different information that is beneficial to each. In this database, we will be analyzing five separate groups: The number of cars designated to each dealership, types of models designated to the dealership, the model of car the dealership sells to the consumer, the price of the car sold, and how many cars are left for sale on the dealership lot. In the first table, "The number of cars designated to each dealership", we will track how many cars the manufactures send to each dealer. This table is important since it will provide information to the manufacturers regarding how many cars, they should send to each dealership in order to maximize their profits. They can also use this information to identify which dealerships are best designated to take in the largest number of cars, which would help them prioritize dealerships in order to make the best use of their resources. In our next table, "Types of models designated to the dealership", we will analyze the different models that each dealership will receive from the manufacturers. This table will help the manufacturer understand which models they should produce more of and what percentage of each type of model they should produce so that dealerships order the proper amount of each model without waste. In addition, our next table, "The model of car the dealership sells to the consumer" will track what model is actually sold to the consumer to help track which models are selling well and which ones aren't. For "The price of the vehicle sold", will track how much each car was sold for so it can be compared to the MSRP price set to the manufacturer. This table determines which models are the most lucrative within their dealerships which identifies their value to the manufacturer, again linking the sales within the dealership to the manufacturer and how they utilize their resources to produce the proper number of models of a car. Lastly, "The number of cars remaining at the dealership" will track how many vehicles are still for sale at the dealership so the manufacturer knows whether or not to supply more cars to the dealership. This table reinforces the model of having the dealerships statistics regarding various aspects of car sales report back to the manufacturer in order to maximize efficiency and use of resources in an effort to achieve maximum profit. We would implement constraints on these tables by not allowing negative values to be inputted into the categories as they would be considered inaccurate results and would not be possible. We would also have to set a range for the allowed inputted data (based on the production scale of the manufacturer) for these tables to block inaccurate results from entering the data and skewing our analysis of the data. The number of cars designated to each dealership would be related to the type of models designated to the dealerships by identifying the optimal amount of particular models that should be designated to the dealership and using that amount to determine the total number of cars that should be designated to each respective dealership. The type of model designated to the dealership relates to the model the dealership sells to the consumer by examining how many of each model is sold to the consumers and therefore depicts an idea of the number of that particular model that should be designated to a dealership for optimized profit. The type of model the dealership sells to the consumer relates to the price of the vehicle sold, by examining the most profitable models and determining how their quantities should be determined compared to the quantities of other models in order to achieve optimal efficiency. The number of cars remaining in the dealership relates back to the cars designated to the dealership and their respective models, by determining which cars were not able to be sold in a given time frame, consequently providing incentive to lower their production and distribution in the next round of designating cars to dealerships.
